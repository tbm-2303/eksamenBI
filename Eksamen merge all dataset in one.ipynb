```python
def merge_dataframes(df1, df2, df3, df4, country_column='Country'):
    # Step 1: Find the common set of countries
    common_countries = set(df1[country_column]) & set(df2[country_column]) & set(df3[country_column]) & set(df4[country_column])

    # Step 2: Select only the rows corresponding to common countries
    df1_common = df1[df1[country_column].isin(common_countries)]
    df2_common = df2[df2[country_column].isin(common_countries)]
    df3_common = df3[df3[country_column].isin(common_countries)]
    df4_common = df4[df4[country_column].isin(common_countries)]

    # Step 3: Merge DataFrames based on the common set of countries
    merged_df = pd.merge(df1_common, df2_common, on=country_column, how='inner')
    merged_df = pd.merge(merged_df, df3_common, on=country_column, how='inner')
    merged_df = pd.merge(merged_df, df4_common, on=country_column, how='inner')

    return merged_df
```
